```python
import matplotlib.pyplot as plt
```

```python
datax = [7,5,8,11,13, 9, 11]
datay = [5, 7, 9, 12, 15, 16, 17]
```

```python
import matplotlib.style
matplotlib.style.use('ggplot')

plt.plot (datax, datay, color='b',linewidth=2.0)
plt.xlabel("Random Number")
plt.ylabel("Random number difference by 2")
plt.show()
```
